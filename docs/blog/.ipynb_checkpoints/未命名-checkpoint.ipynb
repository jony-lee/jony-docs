```python
import glob
```

```python
a = glob.glob("*.md")
```

```python
a = sorted(a)
a
```

```python
content = []
for v in [x for x in a if "2023 04" in x]:
    with open(v,"r") as f:
        content.append("# "+v[:10].replace(" ","-")+f.read()[13:])
```

```python
print("\n---\n".join(content).replace("\n\n","\n"))
```
